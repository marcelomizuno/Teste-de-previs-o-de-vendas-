```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv('Walmart_sales.csv')
```

```python
df = df[['Store', 'Date', 'Weekly_Sales']]
```

```python
df['Date'] = df['Date'].apply(lambda x: '01-' + '-'.join(x.split('-')[1:]))
df = df.groupby(['Store', 'Date']).sum().reset_index()
df['Date'] = pd.to_datetime(df['Date'], format = "%d-%m-%Y")
df = df.rename(columns={'Weekly_Sales': 'Monthly_Sales'})
```

```python
test_data = df.loc[df['Store'] == 1, ['Date', 'Monthly_Sales']]
test_data = test_data.rename(columns={'Date': 'ds','Monthly_Sales': 'y'})
test_data.sort_values(by='ds', inplace = True)
test_data.reset_index(inplace = True)
test_data.drop('index', axis=1, inplace=True)
```

```python
model = Prophet()
model.fit(test_data)
```

```python
future = model.make_future_dataframe(periods=3, freq='MS')
forecast = model.predict(future)
```

```python
result = pd.DataFrame(forecast['yhat'])
result['date'] = forecast['ds'].dt.strftime('%Y-%m-%d')
result['sales'] = test_data['y']
```

```python
result.to_json(orient='records')
```
